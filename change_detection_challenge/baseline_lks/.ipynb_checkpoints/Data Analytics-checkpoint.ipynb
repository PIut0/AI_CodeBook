```python
import os
import cv2
import numpy as np

def count_segmentation_regions(img_id='2015_DMG_1LB_000066.png'):
    base_url = '/scratch/network/mk8574/change_detection_challenge/data/train'
    img_path = os.path.join(base_url, 'y', img_id)
    
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    result_data = {
        'id': img_id,
        'region': img_id.split('_')[1],
        'type': img_id.split('_')[2],
        'class1': 0,
        'class2': 0,
        'class3': 0,
        'class1_bbox': [],
        'class2_bbox': [],
        'class3_bbox': [],
    }

    if img is not None:
        binary_img = np.array(img).astype(np.uint8)

        num_labels, labels = cv2.connectedComponents(binary_img)
        height, width = img.shape
        
        result_data['width'] = width
        result_data['height'] = height
        
        for i in range(1, num_labels):
            pos_list = np.where(labels == i)
            pos_x = pos_list[0][0]
            pos_y = pos_list[1][0]
            region_size = np.sum(labels == i)
            class_label = binary_img[pos_x][pos_y]
            result_data[f'class{class_label}'] += 1
            
            min_x = min(pos_list[0])
            max_x = max(pos_list[0])
            min_y = min(pos_list[1])
            max_y = max(pos_list[1])
            
            bounding_box = {
                'class': class_label,
                'min_x': min_x,
                'max_x': max_x,
                'min_y': min_y,
                'max_y': max_y
            }
            result_data[f'class{class_label}_bbox'].append(bounding_box)

        return result_data
    
    else:
        print('error', img_id)
        return None

y_dataframe = pd.DataFrame()
info = count_segmentation_regions()
pd.Series(info)
y_dataframe = pd.concat([y_dataframe, pd.Series(info).to_frame().T], ignore_index=True)
y_dataframe
```

```python
image_url = '/scratch/network/mk8574/change_detection_challenge/data/train/y'
image_list = glob(os.path.join(image_url, '*.png'))
y_dataframe = pd.DataFrame()

for image in tqdm(image_list):
    try:
        info = count_segmentation_regions(os.path.basename(image))
    except:
        print(image)
    if info:
        y_dataframe = pd.concat([y_dataframe, pd.Series(info).to_frame().T], ignore_index=True)
```

```python
left = {
    'class1': 0,
    'class2': 0,
    'class3': 0
}

right = {
    'class1': 0,
    'class2': 0,
    'class3': 0
}

for data in y_dataframe.itertuples():
    
    for bbox in data.class1_bbox:
        if bbox['min_y'] >= data.width / 2 and bbox['max_y'] >= data.width / 2:
            right['class1'] += 1
        else:
            left['class1'] += 1
            
    for bbox in data.class2_bbox:
        if bbox['min_y'] >= data.width / 2 and bbox['max_y'] >= data.width / 2:
            right['class2'] += 1
        else:
            left['class2'] += 1

    for bbox in data.class3_bbox:
        if bbox['min_y'] >= data.width / 2 and bbox['max_y'] >= data.width / 2:
            right['class3'] += 1
        else:
            left['class3'] += 1

    # break;

print('left:', left)
print('right:', right)
```
