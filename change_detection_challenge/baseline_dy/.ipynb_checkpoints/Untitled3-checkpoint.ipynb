```python
from PIL import Image
import numpy as np
import os
def split_and_swap(image_path):
    # 이미지 열기
    x_original_image = Image.open('/scratch/network/mk8574/change_detection_challenge/data/train/x/'+image_path)
    y_original_image = Image.open('/scratch/network/mk8574/change_detection_challenge/data/train/y/'+image_path)
    
    o = np.array(y_original_image)
    # 이미지의 너비와 높이 구하기
    width, height = y_original_image.size

    # 반으로 나눌 너비 계산
    half_width = width // 2

    vst  = np.zeros((height, width), dtype=np.uint8)

    true_if_12=True
    for i in range(height):
        for j in range(width):
            if o[i][j]==1 and vst[i][j]==0:
                if j>=half_width and o[i][j-half_width]==0:
                    o[i][j]=0
                    vst[i][j]=1
                    o[i][j-half_width]=2  #2
                else:
                    true_if_12=False
                    break
            if o[i][j]==2 and vst[i][j]==0:
                if j<half_width and o[i][j+half_width]==0:
                    o[i][j]=0
                    vst[i][j]=1
                    o[i][j+half_width]=1  #1
                else:
                    true_if_12=False
                    break
        if true_if_12==False:
            break
    if true_if_12:
        new_img = Image.fromarray(o)
        # Save the image as a PNG file
        new_img.save('/scratch/network/mk8574/change_detection_challenge/swapdata/train/y/'+image_path)

    
        # 이미지의 너비와 높이 구하기
        width, height = x_original_image.size

        # 반으로 나눌 너비 계산
        half_width = width // 2

        # 이미지를 반으로 자르기
        left_half = x_original_image.crop((0, 0, half_width, height))
        right_half = x_original_image.crop((half_width, 0, width, height))

        # 이미지를 좌우로 뒤집어서 새로운 이미지 생성
        new_image = Image.new('RGB', (width, height))
        new_image.paste(right_half, (0, 0))
        new_image.paste(left_half, (half_width, 0))

        # 새 이미지 저장
        new_image.save('/scratch/network/mk8574/change_detection_challenge/swapdata/train/x/'+image_path)
    
```

```python
import numpy as np
from PIL import Image
import os
from datetime import datetime
from tqdm import tqdm
import numpy as np
import random, os, sys, torch, cv2, warnings
from glob import glob
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

prj_dir = os.path.dirname('/scratch/network/mk8574/change_detection_challenge/baseline_dy/')

from modules.utils import load_yaml, save_yaml, get_logger
from modules.scalers import get_image_scaler
from modules.datasets import SegDataset
from models.utils import get_model

def colorize_mask(mask, colors):
    # Colorize the mask based on the provided colors
    # `colors` should be a dictionary mapping each class to a BGR color
    colored_mask = np.zeros((mask.shape[0], mask.shape[1], 3), dtype=np.uint8)

    for cls, color in colors.items():
        colored_mask[mask == cls] = color
    
    return colored_mask

def overlay_mask_on_image(image, mask, alpha=0.5):
    # Overlay the mask on the image
    overlay = cv2.addWeighted(image, alpha, mask, 1 - alpha, 0)
    return overlay

# Define colors for each class in BGR format
class_colors = {
    0: [0, 0, 0],        # Class 0: Black
    1: [255, 255, 0],      # Class 1: Lightblue
    2: [0, 255, 255],      # Class 2: Yellow
    3: [255, 0, 255],      # Class 3: Pink
}
```
